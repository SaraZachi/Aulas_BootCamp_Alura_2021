Indica o valor total gasto por Ano/mês, a partir de 2008, em ambiente hospitalar, no SUS, segundo Unidade da Federação

# **01. Aquisição de leitura de dados reais**

```python
import pandas as pd
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html

import numpy as np

import seaborn as sns
sns.set()

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
```

```python
link_dados = 'https://github.com/SaraZachi/BootCamp-2021/blob/main/Aula%201/Data/A151346189_28_143_208.csv?raw=true'
dados = pd.read_csv(link_dados,encoding = 'ISO-8859-1',
                    skiprows = 3, sep=';',skipfooter=12,
                    thousands = ".", decimal =',')
dados
## Erros ocorridos e soluções
# [01] Erro devido ao acento no a (ã) da linguagem portuguesa
# para resolver foi necessário idicar no código uma linguagem padrão a ser utilizada: encoding = 'ISO-8859-1'
# [02] Erro de formatação do arquivo. Nem todas as linhas possuem o separador ';'
# para resolver descartar/ignorar as linhas sem o separador (;) e indicar qual é o separador: skiprows = 3, sep=';',skipfooter=12
# [03] Indicar os separadores decimais e milhar para não dar erro na formatação dos dados: thousands = ".", decimal =',')
```

```python
ax = dados.plot(x='Unidade da Federação', y='2008/Ago', kind='bar', figsize=(10,6))

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
# https://matplotlib.org/stable/gallery/ticks_and_spines/tick-formatters.html

plt.title('Valor gasto por Unidade de Federação em agosto de 2008', fontsize=16, fontweight='bold')
ax.set_xlabel('Unidade da Federação', fontsize=12, fontweight='bold')
ax.set_ylabel('Valor gasto (R$)', fontsize=12, fontweight='bold')

plt.show()
```
