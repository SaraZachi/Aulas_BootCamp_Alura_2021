Para simplificar o número de variantes do referente estudo, serão analizados o número de Coberturas Vacinais dos 5 imunos com maior número total de aplicações nas Unidades Federativas brasileiras. Para determiná-los, vou utilizar a linha das médias por imuno da tabela acima, ordenando os resultados em ordem decrescente.

```python

import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
```

```python
# Importando dataframes
df_imuno = pd.read_csv('https://github.com/SaraZachi/programa-de-imuno-nacional/blob/main/Dataframes/df_imunos.csv?raw=true')
df_pop = pd.read_csv('https://github.com/SaraZachi/programa-de-imuno-nacional/blob/main/Dataframes/dados.csv?raw=true')

# Transformando 'uf' em index
df_imuno = df_imuno.set_index('uf')
df_pop = df_pop.set_index('uf')
```

```python
df =  df_pop.join(df_imuno)
df
```

```python
# Determinando os 5 imunos com maior número de aplicações pela coluna 'mean' que é a média do número de 
# aplicações do imuno pelos estados brasileiros
sv = df_imuno.describe()
sv = sv.T
sv = sv.sort_values('mean', ascending=False)
sv[:5]
```

```python
# Dataframe dos 5 imunos com maior média de aplicação pelas UFs brasileiras
df_principais = df[['regiao','populacao', 'BCG', 'Tríplice Viral  D1', 'Hepatite B', 'Poliomielite', 'DTP']]
df_principais.T
```

No DataFrame acima (df_principais) podemos perceber que o número de pessoas imunizadas por imuno é nítidamente menor que a população total de cada estado. Vamos plotar um gráfico que irá representar isso.

```python
# Palheta de cores
sns.set_palette(sns.color_palette('magma', 5))

# Plot gráficos
sns.barplot(data= df_principais.sort_values(['regiao'], ascending=False), dodge=False, 
            x= 'uf', y='Poliomielite', hue='regiao')
```
