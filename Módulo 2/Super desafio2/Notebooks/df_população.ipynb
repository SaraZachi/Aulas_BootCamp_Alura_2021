# **Importando DataFrame**

Coberturas Vacinais por Imuno segundo Unidade da Federação

```python
url = 'https://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o'

url_html = requests.get(url).text
soup = BeautifulSoup(url_html, "lxml")
data_content = pd.read_html(str(soup.find_all('table')))
dados = pd.DataFrame(data_content[0])
dados.head()
```

# **Explorando e Limpando o Dataframe**

```python
# Removendo colunas indesejadas
dados = dados.drop(columns=['Posição', '% da pop. total', 'País comparável (habitantes)'])
    
# Renomeando as colunas 'Unidade federativa' e 'População'
dados = dados.rename(columns={'Unidade federativa':'uf', 'População':'populacao'})
    
# Transformando 'uf' em index
dados = dados.set_index('uf')
    
# # Removendo espaços em branco em 'população'
dados['populacao'] = dados['populacao'].str.replace(' ', '').astype('int64')  

dados.head()
```

```python
# Salvando o df no diretório como csv
dados.to_csv(r"dados.csv")
```
